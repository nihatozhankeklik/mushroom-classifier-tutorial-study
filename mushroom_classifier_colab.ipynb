```python
from ultralytics import YOLO
import cv2
import os
import random

model = YOLO(r'/content/gdrive/MyDrive/Model/runs/detect/train/weights/best.pt')

yol = "/content/gdrive/MyDrive/rastgele"


files = os.listdir(yol)
files = [f for f in files if f.endswith('.jpg') or f.endswith('.png')]
random.shuffle(files)

i = 0
for j in files:
    img = cv2.imread(os.path.join(yol, j))


    results = model(img, conf = 0.7)


    for result in results:
        boxes = result.boxes
        masks = result.masks
        keypoints = result.keypoints
        probs = result.probs
        obb = result.obb
        result.save(filename=f'/content/gdrive/MyDrive/Model/sonuclar/result{i}.jpg')
    img = cv2.imread('result.jpg')
    i+=1
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image


resim_yolu = '/content/gdrive/MyDrive/Model/sonuclar/result{}.jpg'
resim_sayisi = len(files)
sira_sayisi, sutun_sayisi = 4,5


genislik, yukseklik = 300, 300


birlesik_genislik = sutun_sayisi * genislik
birlesik_yukseklik = sira_sayisi * yukseklik


birlesik_resim = np.zeros((birlesik_yukseklik, birlesik_genislik, 3), dtype=np.uint8)


for i in range(sira_sayisi):
    for j in range(sutun_sayisi):
        indis = i * sutun_sayisi + j
        resim = Image.open(resim_yolu.format(indis))
        resim = resim.resize((genislik, yukseklik))
        resim = np.array(resim)
        birlesik_resim[i * yukseklik : (i + 1) * yukseklik, j * genislik : (j + 1) * genislik, :] = resim
        if indis >= resim_sayisi-1:
          break

    if indis >= resim_sayisi-1:
      break

sonuc_resim = Image.fromarray(birlesik_resim)
sonuc_resim.save('last_result.png')

img = mpimg.imread('/content/gdrive/MyDrive/Model/last_result.png')


plt.figure(figsize=(15, 12))


plt.imshow(img)
plt.axis('off')

plt.show()
```
